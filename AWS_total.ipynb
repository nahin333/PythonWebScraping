```python
def links(topic_links):
  link = []
  children = topic_links.findChildren("a" , recursive=False)
  for child in children:
   return(child['href'])
```

```python
def get_topic_page(url):

  headers ={
            "Content-Type": "text/plain;charset=UTF-8",
            "Referer": "https://www.amazon.com/",
            "sec-ch-ua": '"Microsoft Edge";v="95", "Chromium";v="95", ";Not A Brand";v="99"',
            "sec-ch-ua-mobile": "?0",
            "sec-ch-ua-platform": "Windows",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.44"
        }
  
  response = requests.get(url, headers = headers)
  
  if response.status_code !=200:
    raise Exception('Failed to load page'.forat())
    
  doc = BeautifulSoup(response.text, 'html.parser')
  return doc
```

```python
def get_topic_items(doc):

  topic_rank_tags = doc.find_all('span', {'class': 'zg-badge-text'})
  topic_link_tags = doc.find_all('span', {'class':'aok-inline-block zg-item'})
  topic_name_tags = doc.find_all('div', {'class':'p13n-sc-truncate p13n-sc-line-clamp-2'})
  topic_price_tags = doc.find_all('span', {'class':'a-size-base a-color-price'})
  topic_star_tags = doc.find_all('a', {'class':'a-size-small a-link-normal'})

  topic_dict = {'rank':[],'name':[],'price':[],'star':[],'link':[] }
  base_link = 'https://www.amazon.com'

  for i in range(len(topic_rank_tags)):
    topic_dict['rank'].append(topic_rank_tags[i].text)
    topic_dict['name'].append(topic_name_tags[i].text.strip())
    topic_dict['price'].append(topic_price_tags[i].text)
    topic_dict['star'].append(topic_star_tags[i].text)
    topic_dict['link'].append(base_link + links(topic_link_tags[i]))

  return pd.DataFrame(topic_dict)                            
```

```python
url = "https://www.amazon.com/Best-Sellers-Baby/zgbs/baby-products/"
doc = get_topic_page(url)
get_topic_items(doc)
```
